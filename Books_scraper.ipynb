```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
data=[]
for i in tqdm(range(1,51)):
    link='https://books.toscrape.com/catalogue/page-'+str(i)+'.html'
    res=requests.get(link)
    soup=BeautifulSoup(res.text,'html.parser')

    

    for sp in soup.find_all('li',class_='col-xs-6 col-sm-4 col-md-3 col-lg-3'):
        #print(sp)
        #print('-'*100)
        img_link   =     'https://books.toscrape.com/'+sp.find('img').get('src')
        book_link  =     'https://books.toscrape.com/catalogue/'+sp.find_all('a')[-1].get('href')
        title      =     sp.find_all('a')[-1].get('title')
        rating     =     sp.find('p').get('class')[-1]
        book_price =     sp.find('p',class_='price_color').text[1:]
        stocks     =     sp.find('p',class_='instock availability').text.strip()
        data.append([img_link,book_link,title,rating,book_price,stocks])
        #print(book_link+" | "+title+" | "+img_link+" | "+rating+" | "+book_price+" | "+stocks)
        #break
```

```python
len(data)
```

```python
df=pd.DataFrame(data,columns=['img_link','book_link','title','rating','book_price','stocks'])
```

```python
df.to_csv('Books.csv',index=False)
```

```python
data=[]
for link in tqdm(df['book_link']):
    res=requests.get(link)
    soup=BeautifulSoup(res.text,'html.parser')
    
    typ = soup.find('ul',class_='breadcrumb').find_all('a')[2].text
    
    temp=soup.find('table',class_='table table-striped').find_all('td')
    
    UPC = temp[0].text
    price_x = temp[2].text[2:]
    price_i = temp[3].text[2:]
    tax = temp[4].text[2:]
    qn = temp[5].text
    reviews = temp[6].text
    data.append([typ,UPC,price_x,price_i,tax,qn,reviews])
    
    
```

```python
df=pd.DataFrame(data,columns=['category','UPC','price_e_tax','price_i_tax','tax','quantity','reviews'])
```
